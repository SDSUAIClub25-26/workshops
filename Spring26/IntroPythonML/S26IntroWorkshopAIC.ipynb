Each image of the MNIST dataset is encoded in a 784 dimensional vector, representing a 28 x 28 pixel image. Each pixel has a value between 0 and 255, corresponding to the grey-value of a pixel.<br />

### Define Network Dimensions
We define the input dimension (mnist_dim), a hidden layer dimension (hidden_dim), and the output dimension (output_dim) for our neural network. The hidden dimension is set to one-eighth of the input dimension.

image.png

````python
mnist_dim = X.shape[1] # returns 784
hidden_dim = int(mnist_dim/8) # returns 98
output_dim = len(np.unique(mnist.target)) # returns 10
````

````python
mnist_dim, hidden_dim, output_dim
````

### Build a Simple Neural Network (ClassifierModule)
This `ClassifierModule` defines a basic feed-forward neural network using PyTorch's `nn.Module`. It consists of a hidden layer with ReLU activation, a dropout layer for regularization, and an output layer with a softmax activation for classification.

````python
class ClassifierModule(nn.Module):
    def __init__(
            self,
            input_dim=mnist_dim,
            hidden_dim=hidden_dim,
            output_dim=output_dim,
            dropout=0.5,
    ):
        # initializes base nn.Module class
        super(ClassifierModule, self).__init__()
        self.dropout = nn.Dropout(dropout) # dropout layer

        self.hidden = nn.Linear(input_dim, hidden_dim)
        self.output = nn.Linear(hidden_dim, output_dim)

    def forward(self, X, **kwargs):
        X = F.relu(self.hidden(X))
        X = self.dropout(X)
        X = F.softmax(self.output(X), dim=-1)
        return X
````

### Import Skorch for scikit-learn compatibility
Skorch is a wrapper around PyTorch that allows us to use PyTorch models with a scikit-learn-like API, simplifying training and evaluation.

````python
from skorch import NeuralNetClassifier
````

### Instantiate the Neural Network Classifier
Here, we create an instance of our `ClassifierModule` wrapped by `NeuralNetClassifier` from Skorch. We set hyperparameters like `max_epochs` and `lr` (learning rate), and specify the `device` (CPU or GPU) for training. The default optimizer here is Stochastic Gradient Descent (SGD).

````python
torch.manual_seed(0) # for reproducability

net = NeuralNetClassifier(
    ClassifierModule,
    max_epochs=20,
    lr=0.1,
    device=device,
)
````

### Train the Neural Network
We train the neural network using the `fit` method on our training data (`X_train`, `y_train`). Skorch handles the training loop, including forward and backward passes, and optimizer steps.

````python
net.fit(X_train, y_train);
````

### Build a Convolutional Neural Network (CNN) Model
This `Cnn` class defines a more complex neural network architecture suitable for image classification. It uses convolutional layers (`Conv2d`), pooling (`max_pool2d`), dropout for regularization, and fully connected layers (`Linear`).

**Convolutional 2D Layer (Conv2d)**

Applies small learnable filters that slide across the image to detect local patterns like edges, corners, and textures. Early layers learn simple features, while deeper layers combine them into more complex shapes (like parts of digits). These filters are applied to each training image at different resolutions and the output of each convolved image is input to the next layer.

**Max Pooling (max_pool2d)**

Reduces the spatial size of feature maps by taking the maximum value in small regions. This makes the model more efficient and helps it focus on the most important features while adding some robustness to small shifts in the image.

**Dropout**

Randomly ‚Äúturns off‚Äù a fraction of neurons during training. This prevents the network from relying too heavily on any single feature and helps reduce overfitting, leading to better generalization on unseen data.

**Resources:**
For a nice, visual explanation of CNN's, check out this [page](https://poloclub.github.io/cnn-explainer/) or the visual below

**Think of it like this:**

- Channels = feature detectors

- Spatial dimensions = where those features appear

- More channels = richer vocabulary of features

1. First layer:

‚ÄúWhere are edges and simple strokes?‚Äù

2. Later layers:

‚ÄúWhere are loops, crossings, digit-like shapes?‚Äù

````python
class Cnn(nn.Module):
    def __init__(self, dropout=0.5):
        super(Cnn, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3) # each of 32 output channels represents a feature map with a distinct filter!
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3)
        self.conv2_drop = nn.Dropout2d(p=dropout)
        self.fc1 = nn.Linear(1600, 100) # 1600 = number channels * width * height
        self.fc2 = nn.Linear(100, 10)
        self.fc1_drop = nn.Dropout(p=dropout)

    def forward(self, x):
        x = torch.relu(F.max_pool2d(self.conv1(x), 2))
        x = torch.relu(F.max_pool2d(self.conv2_drop(self.conv2(x)), 2))

        # flatten over channel, height and width = 1600
        x = x.view(-1, x.size(1) * x.size(2) * x.size(3))

        x = torch.relu(self.fc1_drop(self.fc1(x)))
        x = torch.softmax(self.fc2(x), dim=-1)
        return x
````

### Instantiate the CNN Classifier
We create an instance of our `Cnn` model wrapped by `NeuralNetClassifier`. We use a different optimizer (Adam) and adjust hyperparameters for the CNN.

````python
torch.manual_seed(0)

cnn = NeuralNetClassifier(
    Cnn,
    max_epochs=10,
    lr=0.002,
    optimizer=torch.optim.Adam,
    device=device,
)
````

### Train the CNN Model
Train the Convolutional Neural Network using the `fit` method on the CNN-formatted training data.

````python
cnn.fit(XCnn_train, y_train);
````

### Make Predictions with CNN on the Test Set and Evaluate

Use the trained CNN model to make predictions on the test set.

````python
accuracy_score(y_test, y_pred_cnn)
````

To make sure we did it right, we can use Matplotlib's plot feature to see our original image before we preprocess it.

## Preprocessing the Image

ML tasks can get computationally very heavy. We will resize our image to 100x100 pixels to make this quicker. We will then flatten this 2D image array into a single array for the algorithm.

````python
# Resize the image for faster computation
image_resized = img.resize((100, 100))
# Convert the resized image to a NumPy array
image_array = np.array(image_resized)
# Flatten the image into a 2D array (each row represents a pixel)
pixels = image_array.reshape(-1, 3)
````

## Using K-means clustering
We're utlizing a K-means clustering model with 5 clusters (default for this demo, feel free to change it). The algorithm will try to group the pixels into 5 distinct clusters based on their color.

Finding the centroid will give us the average RGB value for each cluster, which is considered the dominant color in the image.

Check out this resource for a great explanation + implementation of how the algorithm works!

````python
# K-Means clustering
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(pixels)
dominant_colors = kmeans.cluster_centers_
````

## Displaying our color pallette
By converting the floating value points into integers, we can represent our colors properly. This plots our colors in a nice colorful bar.

````python
# Ensure the colors are in the correct range for RGB (0-255)
dominant_colors = np.clip(dominant_colors, 0, 255).astype(int)
# Plot the dominant colors
plt.imshow([dominant_colors])  # Plot as a 1xN image (1 row with N dominant colors)
plt.axis('off')  # Cleaner display
plt.show()
````
